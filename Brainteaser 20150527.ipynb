## Solution

```python
#Standard Libaries
%matplotlib inline
import matplotlib.pyplot as plt
import time
from random import random
import numpy as np
```

```python
def Season_Simulator(win_probs=[0.5,0.5,0.5,0.5,0.5], num_games=162, simulations=1000):
    #Start time for simulation
    start_time = time.time()
    
    #Set win probabilities for each of the 5 teams. Default is 50%
    a,b,c,d,e = win_probs[0],win_probs[1],win_probs[2],win_probs[3],win_probs[4]

    #Array of the max number of wins in the season
    season_winner = []
    
    #Run simulation
    for j in range(0,simulations):
        a_wins, b_wins, c_wins, d_wins, e_wins = [],[],[],[],[]
        season=[]
        
        #Simulate a season. Default season has 162 games
        for i in range (0,num_games):
            #Randomly generate numbers between 0-1.
            #If the win probability is greater than the random number, the team wins.
            if a > random(): a_wins.append(1)
            if b > random(): b_wins.append(1)
            if c > random(): c_wins.append(1)
            if d > random(): d_wins.append(1)
            if e > random(): e_wins.append(1)

        #Record the number of wins for each team
        season = [np.sum(a_wins),np.sum(b_wins),np.sum(c_wins),np.sum(d_wins),np.sum(e_wins)]
        
        #The team with the most wins is the season winner
        season_winner.append(max(season))
    
    print("Simulations:        %i" % (simulations))
    print("Average Max Wins:   %.3f" % (np.mean(season_winner)))
    print("Runtime:            %.2f" % (time.time() - start_time))

    #Return the average number of wins for the highest winning team each season.
    return season_winner
```

```python
#Run 1 million simulations
x = Season_Simulator(simulations=1000000)

#Plot a histogram of the means
plt.hist(x,20, normed=0, facecolor='blue', alpha=0.5)
plt.xlabel('Most Wins per Season')
plt.ylabel('Counts')
plt.show()
```

## Further Exploration

```python
#List of iterations to show
iteration_attempts = (1,10,100,1000)

for n in iteration_attempts:
    means = [] #Clear array of means for each set of iterations

    #Run 1,000 examples of n simulations
    for j in range(0,1000):
        z = Season_Simulator(simulations=n)
        means.append(z) #Store the mean of each simulation in the array

    #Plot a histogram of the means
    plt.hist(means,20, normed=0, facecolor='blue', alpha=0.5)
    plt.xlabel('Average Max Wins per Season')
    plt.ylabel('Counts')
    plt.title('HISTOGRAM OF MAX WINS PER SEASON \n Simulations=%i, Mean = %.4f'%(n,np.mean(means)))

    plt.show()
```
